```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import Isomap
from sklearn.lda import LDA
from scipy.io import loadmat
%matplotlib inline
```

```python
X,color = make_swiss_roll(n_samples=3000, noise=0.05, random_state=None)
n_components = 2
print max(color), color.shape
```

```python
# PCA
pca = PCA(n_components)
X_pca = pca.fit_transform(X)
```

```python
# KernelPCA
kpca = KernelPCA(n_components, kernel="rbf", degree=2)
X_kpca = kpca.fit_transform(X)
```

```python
# IsoMap
Y = Isomap(n_neighbors=5, n_components=2)
Y.fit_transform(X)
```

```python
plt.figure()
plt.scatter(X_pca[:, 0], X_pca[:,1], marker='o', c=color)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Cartesian coordinate PCA: ')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
# PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

Y = np.zeros(X_pca.shape[0])
plt.figure()
plt.scatter((X_pca[:,0]),Y, marker='o', c=color)
plt.xlabel('PC1')
plt.title('1D PCA: ')
#plt.xscale('log')
#plt.yscale('log')

plt.axis('tight')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
# KernelPCA
kpca = KernelPCA(n_components=2, kernel="poly", degree=2)
X_kpca = kpca.fit_transform(X)
plt.figure()
plt.scatter(X_kpca[:, 0], X_kpca[:,1], marker='o', c=color)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Cartesian coordinate PCA: ')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
kpca = KernelPCA(n_components=1, kernel="poly", degree=2)
X_kpca = kpca.fit_transform(X)
Y = np.zeros(X_kpca.shape[0])
plt.figure()
plt.scatter(X_kpca[:, 0], Y, marker='x', c=color)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Cartesian coordinate PCA: ')
plt.axis('tight')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
# IsoMap
Y = Isomap(n_neighbors=5, n_components=2, eigen_solver='auto', tol=0, max_iter=None, path_method='D', neighbors_algorithm='auto')
X_imap = Y.fit_transform(X)
```

```python
plt.figure()
plt.scatter(X_imap[:, 0], X_imap[:,1], marker='o', c=color)
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('IsoMap 2D reduction: ')
plt.axis('tight')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
# IsoMap
Y = Isomap(n_neighbors=5, n_components=1, eigen_solver='auto', tol=0, max_iter=None, path_method='D', neighbors_algorithm='auto')
X_imap = Y.fit_transform(X)
Y = np.zeros(X_imap.shape[0])
plt.figure()
plt.scatter(X_imap[:, 0], Y, marker='x', c=color)
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('IsoMap 1D reduction: ')
plt.axis('tight')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
# LDA
y = np.array(color.astype('int'))
lda = LDA(solver='svd', shrinkage=None, priors=None, n_components=2, store_covariance=False, tol=0.0001)
X_lda = lda.fit_transform(X, y)
plt.figure()
plt.scatter(X_lda[:, 0], X_lda[:,1], marker='o', c=color)
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('LDA 2D reduction: ')
plt.axis('tight')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
lda = LDA(solver='svd', shrinkage=None, priors=None, n_components=1, store_covariance=False, tol=0.0001)
X_lda = lda.fit_transform(X, y)
Y = np.zeros(X_lda.shape[0])
plt.figure()
plt.scatter(X_lda[:, 0], Y, marker='x', c=color)
plt.xlabel('F1')
plt.ylabel('0')
plt.title('LDA 2D reduction: ')
plt.axis('tight')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```
