```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import Isomap
%matplotlib inline
```

```python
X,color = make_swiss_roll(n_samples=3000, noise=0.05, random_state=None)
n_components = 2
```

```python
# PCA
pca = PCA(n_components)
X_pca = pca.fit_transform(X)
```

```python
# KernelPCA
kpca = KernelPCA(n_components, kernel="poly", degree=2)
X_kpca = kpca.fit_transform(X)
```

```python
# IsoMap
#Y = Isomap(n_neighbors=10, n_components).fit_transform(X)
```

```python
plt.figure()
plt.scatter(X_pca[:, 0], X_pca[:,1], marker='o', c=color)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Cartesian coordinate PCA: ')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
# PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

Y = np.zeros(X_pca.shape[0])
plt.figure()
plt.scatter((X_pca[:,0]),Y, marker='o', c=color)
plt.xlabel('PC1')
plt.title('1D PCA: ')
#plt.xscale('log')
#plt.yscale('log')

plt.axis('tight')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
# KernelPCA
kpca = KernelPCA(n_components=2, kernel="poly", degree=2)
X_kpca = kpca.fit_transform(X)
plt.figure()
plt.scatter(X_kpca[:, 0], X_kpca[:,1], marker='o', c=color)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Cartesian coordinate PCA: ')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```

```python
kpca = KernelPCA(n_components=1, kernel="poly", degree=2)
X_kpca = kpca.fit_transform(X)
Y = np.zeros(X_kpca.shape[0])
plt.figure()
plt.scatter(X_kpca[:, 0], Y, marker='o', c=color)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Cartesian coordinate PCA: ')
cbar = plt.colorbar()
cbar.set_label('?')
plt.show()
```
